```python
import pandas as pd
import os
import numpy as np
a=['./data/'+i for i in os.listdir('./data/')]
```

```python
columns_to_drop = [
    'src_ip',
    'dst_ip',
    'dst_port',
    'src_port',
    'timestamp',
    'fwd_psh_flags',
    'bwd_psh_flags',
    'fwd_urg_flags',
    'bwd_urg_flags',
    'flow_byts_s', 
    'flow_pkts_s', 
]
```

```python
df_dataset=pd.DataFrame()
for i in a:
    df_dat=pd.read_csv(i)
    if i =='./data/BENIGN.csv' or './data/BENIGN1.csv':
        df_dat['Label']=0
    if i =='./data/http.csv':
        df_dat['Label']=1
    if i =='./data/udp.csv':
        df_dat['Label']=2
    
    df_dataset=df_dataset.append(df_dat,ignore_index=True)
df_dataset = df_dataset.drop(columns=columns_to_drop)
df_dataset
```
